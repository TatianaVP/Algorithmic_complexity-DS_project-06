Напишем функцию, которая будет переводить секунды в более подходящие величины времени:
*   если больше 60 секунд, то выдавать в минутах;
*   если больше 60 минут, то выдавать в часах;
*   если больше 24 часов, то выдавать в днях;
*   функция возвращает число и наименование величины времени: «6.767229629629629 дней»;
*   используйте функцию в каждом из упражнений этого проекта.





```python
def format_time(time):
  if time<=60:
    return f"{time} секунд"
  else:
    time/=60
  if time<=60:
    return f"{time} минут"
  else:
    time/=60
  if time<=24:
    return f"{time} часов"
  else:
    time/=24
    return f"{time} дней"
```

Напишем функцию для расчета времени обработки списка:

```python
def calc_time(f, n_low, time_low, n_high):
  return format_time(time_low * f(n_high) / f(n_low))
```

**Task 2**

O(log(n))

```python
def search(data, value):
    n = len(data)
    left = 0
    right = n - 1
    while left <= right:
        middle = (left + right) // 2
        if value < data[middle]:
            right = middle - 1
        elif value > data[middle]:
            left = middle + 1
        else:
            return middle
    raise ValueError('Value is not in the list')
```

**Task 4**

O(nlog2(n))

```python
def merge(left, right):
    n = len(left)
    m = len(right)
    i = 0
    j = 0
    result = [None] * (n + m) 
    while i + j < n + m:
        if j == m or (i < n and left[i] < right[j]):
            result[i + j] = left[i]
            i += 1
        else:
            result[i + j] = right[j]
            j += 1
            
    return result
 
def sorter(array):
 
    n = len(array)
    if n <= 1:
        return array
 
    middle = int(n // 2)
    left = sorter(array[:middle])
    right = sorter(array[middle:])
 
    return merge(left, right)
```

**Task 5**

O(n^2)

```python
def sorter(data):
    swapped = True
    while swapped:
        swapped = False
        for i in range(len(data)-1):
            if data[i] > data[i+1]:
                data[i], data[i+1] = data[i+1], data[i]
                swapped = True
```

**Task 7**

```python
def permutation(data, n):
    if n == 1:
        print(data)
        return
    
    for i in range(n):
        permutation(data, n - 1)
        if n % 2 == 0:
            data[i], data[n-1] = data[n-1], data[i]
        else:
            data[0], data[n-1] = data[n-1], data[0]
```

```python
def my_func(data):
    i = 0
    j = 0
    m = 0
    c = 0
    while i < len(data):
        if data[i] == data[j]:
            c = c + 1
        j = j + 1
        if j >= len(data):
            if c > m:
                m = c
            c = 0
            i = i + 1
            j = i
    return m
```

Функция считает, сколько повторов.

У данного кода алгоритмическая сложность - O(n^2).

Перепишем код ниже так, чтобы его алгоритмическая сложность была ниже, но при этом так, чтобы он по-прежнему выполнял ту же задачу.

```python
def my_func_new(data):
    dict = {}
    m = 0
    for i in data:
      if i not in dict:
        dict[i] = 0
      dict[i] += 1
      if dict[i] > m:
        m = dict[i]
    return m
```

У нового кода алгоритмическая сложность - O(n).
